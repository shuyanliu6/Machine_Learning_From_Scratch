```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Salary_dataset.csv')
df.head(5)
```

```python
from main import LinearRegression
```

```python
x = df.YearsExperience
y = df.Salary
```

```python
model = LinearRegression(learning_rate=0.001,epochs=10000)

model.fit(x,y)
b0,b1 = model.get_params()
b0,b1
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color='black')
plt.plot(list(range(0,20)), [b1 * x + b0 for x in range(0,20)], color='red')
plt.show()
```

```python
## comparing with scikitlearn
from sklearn.linear_model import LinearRegression as SklearnLinearRegression
x1 = np.array(x).reshape(-1,1)
sklearn_model = SklearnLinearRegression()
sklearn_model.fit(x1, y)

sklearn_b0 = sklearn_model.intercept_
sklearn_b1 = sklearn_model.coef_[0]
sklearn_b0,sklearn_b1
```

```python
plt.scatter(x,y, color='black')
plt.plot(list(range(0,20)), [sklearn_b1 * x + sklearn_b0 for x in range(0,20)], color='red')
plt.show()
```
